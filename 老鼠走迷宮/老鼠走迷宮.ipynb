# 老鼠走迷宮

老鼠走迷宮是經典程式設計範例，這個遊戲牽涉到以下幾個技巧與概念：

    1.讀檔功能：因為迷宮的由人工輸入實在太感人了，所以通常以讀檔來處理迷宮地圖。
    2.二維陣列：用以定位迷宮的特徵與老鼠的位置。
    3.Stack堆疊：已經走訪過的路徑需要存在堆疊中，遇到走不通需回頭時，要將儲存的路徑以後進先出的方式，回頭找路。
    4.DFS(Depth First Search)深度優先搜尋走訪迷宮：先選擇一條路徑走到底，直到不能走之後再回頭走另一條。
    5.Queue佇列：走到終點之後，要將沿途的風景以先進先出的方式印出來。
